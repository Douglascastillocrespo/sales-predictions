##5) Decidan cómo abordar los valores faltantes y cómo hacerlo. (Esto requiere sus criterios, así que expliquen su elección).

```python
#llenaremos los valores faltantes en la columna 'Item_Weight' con la mediana por ser numerica
df['Item_Weight'] = df['Item_Weight'].fillna(df['Item_Weight'].median())

#llenaremos los valores faltantes en la columna 'Outlet_Size' con la el valor mas frecuente (moda) por ser categorica
df['Outlet_Size'] = df['Outlet_Size'].fillna(df['Outlet_Size'].mode()[0])

df
```
